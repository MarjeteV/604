```python
#What is the number of occurrences of rolling two 6s simultaneously with 1000 simulations?
import matplotlib.pyplot as plt

def roll_dice():
    return random.randint(1, 6)

def simulate_game(num_simulations):
    occurrences = []
    total_occurrences = 0
    
    for _ in range(num_simulations):
        roll1 = roll_dice()
        roll2 = roll_dice()
        
        if roll1 == 6 and roll2 == 6:
            total_occurrences += 1
        
        occurrences.append(total_occurrences)
    
    return occurrences

num_simulations = 1000
simulations = simulate_game(num_simulations)

# Plotting the time series
plt.plot(range(1, num_simulations + 1), simulations)

plt.xlabel('Simulation Step')
plt.ylabel('Occurrences')
plt.title('Time Series of Rolling Two 6s')

plt.show()
```
